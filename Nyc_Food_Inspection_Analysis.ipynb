```python
import pandas as pd

# Convert IDs to strings
df['camis'] = df['camis'].astype(str)

# Restaurant name & categorical text fields to string
string_cols = [
    'dba', 'boro', 'zipcode', 'cuisine_description',
    'action', 'violation_code', 'violation_description',
    'critical_flag', 'grade', 'inspection_type'
]

for col in string_cols:
    df[col] = df[col].astype(str)

# ZIPCODE must NOT be float
df['zipcode'] = df['zipcode'].str.replace('.0','', regex=False)

# Convert date columns to datetime
df['inspection_date'] = pd.to_datetime(df['inspection_date'], errors='coerce')
df['grade_date'] = pd.to_datetime(df['grade_date'], errors='coerce')
df['record_date'] = pd.to_datetime(df['record_date'], errors='coerce')

# Convert numeric fields
df['score'] = pd.to_numeric(df['score'], errors='coerce')

df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
```

```python
# 1. Convert to datetime
date_cols = ['inspection_date', 'grade_date', 'record_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# 2. Extract date features
df['inspection_year'] = df['inspection_date'].dt.year
df['inspection_month'] = df['inspection_date'].dt.month
df['inspection_day'] = df['inspection_date'].dt.day
df['inspection_weekday'] = df['inspection_date'].dt.day_name()
df['inspection_quarter'] = df['inspection_date'].dt.quarter

# 3. Remove future inspection & grade dates
today = pd.Timestamp.today()
df.loc[df['inspection_date'] > today, 'inspection_date'] = pd.NaT
df.loc[df['grade_date'] > today, 'grade_date'] = pd.NaT

# 4. Detect negative time intervals
df['grade_delay_days'] = (df['grade_date'] - df['inspection_date']).dt.days
df['record_delay_days'] = (df['record_date'] - df['inspection_date']).dt.days

# Fix negative intervals
df.loc[df['grade_delay_days'] < 0, 'grade_date'] = pd.NaT
df.loc[df['record_delay_days'] < 0, 'record_date'] = pd.NaT

# 5. Sort data
df = df.sort_values('inspection_date')
```
